```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Weather-2020-2025.csv")
```

```python
#Quick look into the data
data.info()
data.head(5)
```

```python
#Remove null values
data.dropna()
```

```python
#Max temp group
data.groupby("location_id")["temperature_2m"].max()
```

```python
#Min temp group 
data.groupby("location_id")["temperature_2m"].min()
```

```python
#Understanding the location of location 0
data.loc[data['temperature_2m'] == 33.5, ['time']]
```

```python
#Understanding the location of location 1
data.loc[data['temperature_2m'] == 26.6, ['time']]
```

```python
#Understanding the location of location 2
data.loc[data['temperature_2m'] == 40.7 , ['time']]
```

```python
#Find entries per location
data.groupby("location_id").size()
```

```python
#Reduce data frame to entries during the daytime
dayFilter = data[data["is_day"] == 1]
dayFilter
```

```python
#Find how many entries have data during the day
dayFilter.groupby("location_id").size()
```

```python
#Solar power works best when temp < 23°C
dayTempFilter = dayFilter[dayFilter['temperature_2m'] <23]
dayTempFilter
```

```python
#Check how many entries are at the right time of day and temp
dayTempFilter.groupby("location_id").size()
```

```python
#new dataframe that accounts for cloud cover
dayTempCloudFilter = dayTempFilter[dayTempFilter['cloud_cover (%)'] < 50]
dayTempCloudFilter
```

```python
#Find the size of the dataframe that filers temp, cloud and daytime
dayTempCloudFilter.groupby("location_id").size()
```
