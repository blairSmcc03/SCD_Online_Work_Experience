## Exercise - Update the file info.txt with your own information (Students only)

We want to amend a file someone else created and track our changes in our own branch, and push those changes to the origin repository.

```python
# Create your own branch for your work and switch to it
git branch <username_branch>
git checkout <username_branch>
# Edit the file with the vim text editor
vim info.txt # Update with your own info
:wq #save the changes and quit the editor
# Observe the changes
git diff info.txt
# Stage the changes
git add info.txt
# Commit those changes to your branch with appropriate commit message
git commit -m "Update user information file"
# Observe the new status
git status
# Push changs to your own branch on the origin repo
git push origin <username_branch>
```

## Demo: Fetching files from the origin

At times we will need to update our local repository with changes made to the origin.

We can pull / fetch and merge changes from the origin

```python
# Use git log to observe the most recent commit on current branch
git log -1 
# Show the current files in the directory
ls -l
# Download any updated files from the origin. Note: This does not merge them yet!
git fetch origin 
# Confirm no changes made by checking no change to the most recent commit
git log -1 to show not merged to local yet
# Merge the files into local branch. This will commit them
git merge origin main
# Observe the new latest commit
git log -1 
# Observe new files in local repo
ls -l
```
